## Label Encoding


```python
genders = ['male', 'female']

label_encoder = tf.keras.layers.StringLookup(vocabulary=genders, mask_token=None)

encoded = label_encoder(data.sex)
```

```python
smoke = ['yes', 'no']

label_encoder = tf.keras.layers.StringLookup(vocabulary=smoke, mask_token=None)

encoded2 = label_encoder(data.smoker)
```

```python
regions = data.region.unique()

label_encoder = tf.keras.layers.StringLookup(vocabulary=regions, mask_token=None)

encoded3 = label_encoder(data.region)
```

```python
data['sex'] = encoded
```

```python
data['smoker'] = encoded2
```

```python
data['region'] = encoded3
```

```python
data.head(25)
```

## Normalization

```python
data.age = data.age/tf.reduce_max(data.age)
```

```python
data.bmi = data.bmi/tf.reduce_max(data.bmi)
```

```python
data.head()
```

### Performing the train-test split

```python
from sklearn.model_selection import train_test_split

x = data.iloc[:, :-1]
y = data.iloc[:, -1]

x.head(), y.head()

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

### Defining the Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential([
    Dense(6),
    Dense(15),
    Dense(5),
    Dense(1)
])
```

```python
model.compile(loss = 'mae', optimizer = tf.keras.optimizers.Adam(lr = 0.001), metrics = 'mae')
```

```python
model.fit(x_train, y_train, epochs = 25)
```

```python
model.fit(x_test, y_test)
```
